### **1. Определить (с обоснованием), зависимы или независимы следующие события:**

(a) Несовместные события  

(b) События, образующие $\sigma$ -алгебру $\Sigma$ в пространстве $(\Omega, \Sigma, P)$

(c) События, имеющие одинаковую вероятность

**Решение:**

**(a)**

По определению несовместных событий - P(AB) = 0, а по определению независимых - P(AB) = P(A)P(B).

То есть если несовместные события независимы, то P(A)P(B) = 0, что подразумевает под собой нулевую вероятность как минимум одного из событий - A или B.

В противном случае события зависимы.


**(b)**

Рассмотрим два примера:

В пространстве $\Omega = \{1,2,3,4\}$ с равномерной мерой возьмём $\Sigma = \{\varnothing, \Omega, \{1,2\}, \{3,4\}\}$. События $A = \{1,2\}$ и $B = \{3,4\}$ принадлежат $\Sigma$, но $P(A \cap B) = 0$, а $P(A)P(B) = 1/4$, следовательно, они зависимы.

В то же время в вероятностном пространстве для бросания двух монет $\Omega = \{ГГ, ГО, ОГ, ОО\}$ с равномерной вероятностью $P = 1/4$ на каждом исходе $\sigma$-алгебра $\Sigma$ содержит все подмножества $\Omega$. События $A = \{ОГ, ОО\}$ («орёл на первой монете») и $B = \{ГО, ОО\}$ («орёл на второй монете») принадлежат $\Sigma$, причём $P(A) = P(B) = 1/2$, а $A \cap B = \{ОО\}$, $P(A \cap B) = 1/4 = P(A)P(B)$, значит, они независимы.

Таким образом $\sigma$-алгебра гарантирует только то, что выбранные события допустимы. Она может содержать и зависимые, и независимые события — всё определяется мерой $P$ и зависит от конкретной $\sigma$-алгебры.

**(с)**

Рассмотрим пример, аналогичный примеру из пункта **(b)**:

Пусть подбросили две монеты. События $A = \{ОГ, ОО\}$ («орёл на первой монете») и $B = \{ГО, ОО\}$ («орёл на второй монете»). $P(A)$ и $P(B)$ равны $1/2$, а совместное событие имеет вероятность $1/4 = P(A)P(B)$, то есть события независимы.

В то же время можно рассмотреть следующий пример:

Пусть имеется честная игральная кость. События $A = \{1,2,3\}$ и $B = \{2,3,4\}$ равновероятны ($P = 1/2$), но $P(A \cap B) = 2/6 = 1/3 \neq 1/4 = P(A)P(B)$, значит, они зависимы.

Из этого следует сделать вывод, что одинаковая вероятность не является основой для суждения о зависимости или независимости событий.

* $B$ — появление решки на первой монете;
* $C$ — появление герба на второй монете;
* $D$ — появление решки на второй монете;
* $E$ — появление хотя бы одного герба;
* $F$ — появление хотя бы одной решки;
* $G$ — появление одного герба и одной решки;
* $H$ — непоявление ни одного герба;
* $K$ — появление двух гербов.


Определить, каким событиям этого списка равносильны следующие события:

**(a)** $A + C$

**(b)** $AC$

**(c)** $EF$

**(d)** $G + E$

**(e)** $GE$

**(f)** $BD$

**(g)** $E + K$.

**Решение:**
Пусть выпадение герба обозначается как O, решки - как P

Трансформируем события A-K в результаты выпадения

* $A$ $= \{OP, OO\}$
* $B$ $= \{PO, PP\}$
* $C$ $= \{PO, OO\}$
* $D$ $= \{OP, PP\}$
* $E$ $= \{OP, PO, OO\}$
* $F$ $= \{PO, OP, PP\}$
* $G$ $= \{OP, PO\}$
* $H$ $= \{PP\}$
* $K$ $= \{OO\}$



**(a)** $A + C = A \cup C = \{OP, OO\} \cup \{PO, OO\} = \{OP, PO, OO\} = E$  
**(b)** $AC = A \cap C = \{OP, OO\} \cap \{PO, OO\} = \{OO\} = K$  
**(c)** $EF = E \cap F = \{OP, PO, OO\} \cap \{PO, OP, PP\} = \{OP, PO\} = G$  
**(d)** $G + E = G \cup E = \{OP, PO\} \cup \{OP, PO, OO\} = \{OP, PO, OO\} = E$  
**(e)** $GE = G \cap E = \{OP, PO\} \cap \{OP, PO, OO\} = \{OP, PO\} = G$  
**(f)** $BD = B \cap D = \{PO, PP\} \cap \{OP, PP\} = \{PP\} = H$  
**(g)** $E + K = E \cup K = \{OP, PO, OO\} \cup \{OO\} = \{OP, PO, OO\} = E$

\begin{cases}

\end{cases}

### **5. Самолет, по которому ведется стрельба**

состоит из трех различных по
уязвимости частей:

**(a)** Кабина летчика и двигатель

**(b)** Топливные баки

**(c)** Планер

Для поражения самолета достаточно либо одного попадания в первую
часть, либо двух попаданий во вторую, либо трех в третью. При попа-
дании в самолет одного снаряда, снаряд с вероятностью $P_1$ попадает в
первую часть, с вероятностью $P_2$ — во вторую, с вероятностью $P_3$ — в
третью. Попавшие снаряды распределяются по частям независимо друг
от друга.

Известно, что в самолет попало $m$ снарядов. Найти условную вероят-
ность $P(A|m)$ события $A$ – «Самолет поражен» – при $m = 1, 2, 3, 4$

**Решение**

Самолёт поражён , если выполнено хотя бы одно из условий:  

* $X_1 \ge 1$ (хотя бы одно попадание в часть a)  
* $X_2 \ge 2$ (не менее двух в часть b)  
* $X_3 \ge 3$ (не менее трёх в часть c)

То есть

Самолёт не поражён, если:  
$X_1 = 0$, $X_2 \le 1$, $X_3 \le 2$


**При $m = 1$:**  
Поражение только при попадании в часть a:  
$$
P(A \mid 1) = p_1 = 1 - p_2 - p_3
$$

**При $m = 2$:**  
Условие $X_3 \le 2$ выполнено автоматически  

Исходы, при которых поражения самолет нет:  
- $(0,0,2)$: вероятность $p_3^2$  
- $(0,1,1)$: вероятность $2 p_2 p_3$  
    
$$
P(A \mid 2) = 1 - p_3^2 - 2 p_2 p_3
$$


**При $m = 3$:**  
Исходы, при которых поражения самолета нет:  
- $(0,1,2)$: вероятность $3 p_2 p_3^2$  

$$
P(A \mid 3) = 1 - 3 p_2 p_3^2
$$

**Случай $m = 4$:**  
Максимальное количество попаданий, которое может выдержать самолет равняется 3. В противном случае по прицнипу дирихле, хотя бы одна из его частей будет критически повреждена. То есть условня вероятность равна 1.

### 1. Пусть  
$$
f_\xi(x,y) = \frac{e^{-2|y|}}{\pi(1+x^2)}
$$  
Является ли данная функция плотностью распределения случайного вектора?



**Решение:**

Проверим функцию по следующим пунктам:

**Неотрицательность:**  
Функция $e^{-2|y|} > 0$ при всех $y$,  
$1 + x^2 > 0$ при всех $x$,  
константа $\pi > 0$.  

Значит, $f_\xi(x,y) \ge 0$ для всех $(x,y) \in \mathbb{R}^2$.  
Первое условие плотности выполнено.

**Нормировка на 1:**  
Проверим интеграл по $\mathbb{R}^2$:  
$$
\iint_{\mathbb{R}^2} f_\xi(x,y) \,dx\,dy = \int_{-\infty}^{\infty} \left[ \frac{1}{\pi(1+x^2)} \right] dx \cdot \int_{-\infty}^{\infty} e^{-2|y|} \,dy
$$  
Первый интеграл:  
$$
\int_{-\infty}^{\infty} \frac{dx}{1+x^2} = \pi \quad \Rightarrow \quad \frac{1}{\pi} \cdot \pi = 1
$$  
Второй интеграл:  
$$
\int_{-\infty}^{\infty} e^{-2|y|} \,dy = 2 \int_0^{\infty} e^{-2y} \,dy = 2 \cdot \frac{1}{2} = 1
$$  
Произведение равняется единице, следовательно условие нормировки выполнено.

**Вывод:**  
Оба условия для плотности распределения выполнены:  
$f(x,y) \ge 0$  
$\iint_{\mathbb{R}^2} f(x,y) \,dx\,dy = 1$  

**Ответ: Да, является.**

### 4. Пусть $\xi \sim U_{-\pi,\pi}$ и $\eta_1 = \cos \xi$, $\eta_2 = \sin \xi$.  
(a) Вычислить математическое ожидание, ковариационную и корреляционную матрицы вектора $(\eta_1, \eta_2)$  
(b) Исследовать $\eta_1$ и $\eta_2$ на независимость и некоррелированность

**Решение:**

**(a)**  
**Математические ожидания $\eta_1$ и $\eta_2$:**  
$$
E[\eta_1] = \int_{-\pi}^{\pi} \cos x \cdot \frac{1}{2\pi} \, dx = 0
$$  
$$
E[\eta_2] = \int_{-\pi}^{\pi} \sin x \cdot \frac{1}{2\pi} \, dx = 0
$$  

Дисперсии $\eta_1$ и $\eta_2$:  
$$
E[\eta_1^2] = \int_{-\pi}^{\pi} \cos^2 x \cdot \frac{1}{2\pi} \, dx = \frac{1}{2\pi} \int_{-\pi}^{\pi} \frac{1 + \cos 2x}{2} \, dx = \frac{1}{2\pi} \left[ \pi + 0 \right] = \frac{1}{2}
$$  
$$
E[\eta_2^2] = \int_{-\pi}^{\pi} \sin^2 x \cdot \frac{1}{2\pi} \, dx = \frac{1}{2\pi} \int_{-\pi}^{\pi} \frac{1 - \cos 2x}{2} \, dx = \frac{1}{2\pi} \left[ \pi - 0 \right] = \frac{1}{2}
$$  
$$
D[\eta_1] = E[\eta_1^2] - (E[\eta_1])^2 = \frac{1}{2} - 0 = \frac{1}{2}
$$  
$$
D[\eta_2] = E[\eta_2^2] - (E[\eta_2])^2 = \frac{1}{2} - 0 = \frac{1}{2}
$$  

Ковариация $\eta_1$ и $\eta_2$:  
$$
E[\eta_1 \eta_2] = \int_{-\pi}^{\pi} \cos x \sin x \cdot \frac{1}{2\pi} \, dx = \frac{1}{2\pi} \int_{-\pi}^{\pi} \frac{\sin 2x}{2} \, dx = 0
$$  
$$
\operatorname{cov}(\eta_1, \eta_2) = E[\eta_1 \eta_2] - E[\eta_1] E[\eta_2] = 0 - 0 = 0
$$  

**Ковариационная матрица вектора $(\eta_1, \eta_2)$:**
$$
C = \begin{pmatrix} D[\eta_1] & \operatorname{cov}(\eta_1, \eta_2) \\ \operatorname{cov}(\eta_1, \eta_2) & D[\eta_2] \end{pmatrix} = \begin{pmatrix} 1/2 & 0 \\ 0 & 1/2 \end{pmatrix}
$$  

**Корреляционная матрица:**  
$$
\rho = \frac{\operatorname{cov}(\eta_1, \eta_2)}{\sqrt{D[\eta_1] D[\eta_2]}} = \frac{0}{\sqrt{(1/2)(1/2)}} = 0
$$  
$$
R = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}
$$  

**(b)**  


**Некоррелированность:**  
Поскольку $\operatorname{cov}(\eta_1, \eta_2) = 0$, случайные величины $\eta_1$ и $\eta_2$ являются **некоррелированными**.  

**Независимость:**  
Заметим, что:  
$$
\eta_1^2 + \eta_2^2 = \cos^2 \xi + \sin^2 \xi = 1
$$  
→ $\eta_1$ и $\eta_2$ лежат на единичной окружности.  
Вектор $(\eta_1, \eta_2) = (\cos \xi, \sin \xi)$ — детерминированная функция от $\xi$.  
Таким образом, $\eta_1$ и $\eta_2$ функционально зависимы → не могут быть независимыми.  

Формально:  
$P(\eta_1^2 + \eta_2^2 = 1) = 1$, но если $\eta_1$ и $\eta_2$ независимы, то $\eta_1^2 + \eta_2^2$ должно принимать разные значения с положительной вероятностью $⇒$ противоречие.

### 5. Найти плотность распределения суммы двух независимых случайных величин $\xi$ и $\eta$, если $\xi \sim \operatorname{Exp}(2)$ и $\eta \sim U[0,1]$.

**Решение:**

Плотность распределения случайной величины $\xi$ имеет вид
$$f_\xi(x) =
\begin{cases}
2e^{-2x}, & x \geqslant 0, \\
0, & x < 0.
\end{cases}$$
Плотность распределения случайной величины $\eta$ равна
$$f_\eta(y) =
\begin{cases}
1, & 0 \leqslant y \leqslant 1, \\
0, & \text{иначе}.
\end{cases}$$
Пусть $\zeta = \xi + \eta$. Поскольку $\xi$ и $\eta$ независимы, плотность распределения суммы находится по формуле свёртки:
$$f_\zeta(z) = \int_{-\infty}^{+\infty} f_\xi(x) f_\eta(z-x) \, dx.$$
Область значений $\zeta$: $\xi \geqslant 0$, $\eta \in [0,1]$, следовательно, $\zeta \geqslant 0$. Рассмотрим случаи по $z$.

При $z < 0$ интеграл равен нулю, т.е. $f_\zeta(z) = 0$.
При $0 \leqslant z < 1$ условие $z-x \in [0,1]$ выполняется при $x \in [0,z]$. Тогда

$$f_\zeta(z) = \int_{0}^{z} 2e^{-2x} \cdot 1 \, dx = 2 \left[ -\frac{1}{2} e^{-2x} \right]_{0}^{z} = 1 - e^{-2z}.$$

При $z \geqslant 1$ условие $z-x \in [0,1]$ выполняется при $x \in [z-1,z]$. Тогда

$$f_\zeta(z) = \int_{z-1}^{z} 2e^{-2x} \cdot 1 \, dx = \left[ -e^{-2x} \right]_{z-1}^{z} = -e^{-2z} + e^{-2(z-1)} = e^{2} e^{-2z} - e^{-2z} = (e^{2} - 1) e^{-2z}.$$
Итак, плотность распределения суммы $\zeta = \xi + \eta$ имеет вид
$$\boxed{
f_\zeta(z) =
\begin{cases}
0, & z < 0, \\[6pt]
1 - e^{-2z}, & 0 \leqslant z < 1, \\[6pt]
(e^{2} - 1) e^{-2z}, & z \geqslant 1.
\end{cases}
}$$
